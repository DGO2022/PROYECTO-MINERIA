```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

plt.style.use('classic')  # Esto configura el estilo clásico para los gráficos

# Cambia la ruta a la ubicación correcta de tu imagen
imagen = Image.open("./imagenes/frida.jpg")  # Asegúrate de que 'imagenes' está en el directorio correcto


plt.imshow(imagen)  # Muestra la imagen
plt.show()  # Muestra la imagen
```

```python
imagen_gr=imagen.convert('LA')

imagen_mat = np.array(list(imagen_gr.getdata(band=0)), float)

imagen_mat.shape = (imagen_gr.size[1], imagen_gr.size[0])

U, D, V = np.linalg.svd(imagen_mat)

print(imagen_mat.shape)
print(U.shape)
print(D.shape)
print(V.shape)
```

```python
imagen_recons = np.matrix(U[:, :1]) * np.diag(D[:1]) * np.matrix(V[:1,:])
plt.imshow(imagen_recons, cmap='gray')  # Muestra la imagen
plt.show() 
```

```python
i = 5
imagen_recons = np.matrix(U[:, :i]) * np.diag(D[:i]) * np.matrix(V[:i,:])
plt.imshow(imagen_recons, cmap='gray')# Muestra la imagen

titulo = "valores singulares = %s" % i

plt.title(titulo)
plt.show()
```
