```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

plt.style.use('classic')  # Esto configura el estilo clsico para los grficos

# Cambia la ruta a la ubicacin correcta de tu imagen
imagen = Image.open("./imagenes/frida.jpg")  # Asegrate de que 'imagenes' est en el directorio correcto


plt.imshow(imagen)  # Muestra la imagen
plt.show()  # Muestra la imagen
```

```python
imagen_gr=imagen.convert('LA')
print(imagen_gr)
```

```python
imagen_mat =np.array(list(imagen_gr.getdata(band=0)), float)
print(imagen_mat)
```

```python
imagen_mat.shape = (imagen_gr.size[1], imagen_gr.size[0])
print(imagen_mat)
```

```python
print(imagen_mat.shape)
```

```python
plt.imshow(imagen_mat, cmap = 'gray')
plt.show()
```

```python
imagen_mat_2 = imagen_mat /10
print(imagen_mat_2)
```

```python
plt.imshow(imagen_mat_2)
plt.show()
```

```python
plt.imshow(imagen_mat_2, cmap='gray')
plt.show()
```
