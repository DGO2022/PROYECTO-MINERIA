## Fase 3. Data Preparation

### 🧹 4. Data cleaning

#### 4.1 Selección inicial de variables claves

Queremos construir un modelo que prediga si un vehículo será APPROVED (1) o DISAPPROVED (0) sin usar mediciones de gases contaminantes.
Por lo tanto, debemos excluir columnas de emisiones directas y quedarnos solo con las características intrínsecas del vehículo.


a) Candidatas a ELIMINAR (administrativas, metadatos, fechas o con riesgo de leakage)

fecha_emision → administrativa, no aporta a predecir.

fecha_corte → administrativa.

id → identificador único, sin valor predictivo.

fe_inventario → fecha, podría correlacionarse con la salida, pero es administrativa.

fe_proceso → también fecha administrativa.

cantidad_mb_lim → derivada directamente de cantidad_mb (redundante).

b) Candidatas a CONSERVAR (predictores útiles)

producto → tipo de combustible/producto (ej. "G Premium").

cantidad_mb → cantidad medida (numérica, continua).

vlmen_fndo_ttal → volumen total, variable técnica importante.

ANON_CO_OSINERG → código anonimizado (puede representar centro/proveedor).

ANON_PTA_ABASTECIMIENTO → punto de abastecimiento, podría influir en la distribución.

c) Variable objetivo (target)

En tu dataset no tienes algo como "State (APPROVED/DISAPPROVED)".
Pero sí podrías definir una etiqueta objetivo dependiendo de lo que quieras predecir:

Ejemplo 1: Predecir si la cantidad fue “0” o “>0” → variable binaria (similar a aprobado/desaprobado).

Ejemplo 2: Predecir rangos de vlmen_fndo_ttal (bajo, medio, alto).

Ejemplo 3: Si tienes en otra tabla un campo de “estado” o “validación”, usarlo como State.

```python
# ----------------------------------------------
# 📌 Preparación del dataset para minería de datos
# ----------------------------------------------

import pandas as pd

# Cargar el archivo Excel
file_path = "Diecismuenvemil.xlsx"
data = pd.read_excel(file_path, sheet_name="Sheet1")

# ==============================
# a) Candidatas a ELIMINAR
# ==============================
eliminar = [
    'fecha_emision',    # administrativa
    'fecha_corte',      # administrativa
    'id',               # identificador único
    'fe_inventario',    # fecha administrativa
    'fe_proceso',       # fecha administrativa
    'cantidad_mb_lim'   # redundante (derivada de cantidad_mb)
]

# ==============================
# b) Candidatas a CONSERVAR
# ==============================
conservar = [
    'producto',                # tipo de combustible
    'cantidad_mb',             # cantidad medida
    'vlmen_fndo_ttal',         # volumen total
    'ANON_CO_OSINERG',         # código anonimizado (centro/proveedor)
    'ANON_PTA_ABASTECIMIENTO'  # punto de abastecimiento
]

# ==============================
# c) Variable objetivo (target)
# ==============================
# Definimos como ejemplo: si cantidad_mb > 0 → 1 (APROBADO), si =0 → 0 (DESAPROBADO)
data['target'] = (data['cantidad_mb'] > 0).astype(int)

# ==============================
# Crear dataset final para modelado
# ==============================
df_modelo = data[conservar + ['target']]

print("Variables eliminadas:", eliminar)
print("Variables conservadas:", conservar)
print("\nVista previa del dataset preparado:")
df_modelo.head()
```

```python
# ==============================
# Selección de variables relevantes + objetivo
# ==============================

# Definimos las columnas a conservar
features_keep = [
    "producto",
    "cantidad_mb",
    "vlmen_fndo_ttal",
    "ANON_CO_OSINERG",
    "ANON_PTA_ABASTECIMIENTO"
]

# Variable objetivo
target = ["target"]

# Crear nuevo dataframe solo con las columnas seleccionadas
df_sel = data[features_keep + target]

print("df_sel.shape:", df_sel.shape)
print("df_sel.columns:", df_sel.columns.tolist())

# Mostrar primeras filas
df_sel.head(3)
```

#### 📊 Interpretación de correlaciones

##### 1. Variables con correlación positiva (asociadas a mayor `cantidad_mb`)
- **Ejemplo: `vlmen_fndo_ttal` (r ≈ +0.xx):** cuando el volumen de fondo es mayor, también aumenta la `cantidad_mb`.  
- **Ejemplo: `cantidad_mb_lim` (r ≈ +0.xx):** correlación positiva indicando que un límite más alto está asociado con mayor cantidad.

##### 2. Variables con correlación negativa (asociadas a menor `cantidad_mb`)
- **Ejemplo: `fe_proceso` (r ≈ -0.xx):** fechas de proceso más recientes/altas se asocian a menor cantidad.  
- **Ejemplo: `fe_inventario` (r ≈ -0.xx):** posible efecto temporal negativo.

##### 3. Variables neutras o con efecto muy bajo
- Variables con correlación cercana a **0** muestran poca relación directa con la `cantidad_mb`.

---

##### 📌 Conclusiones
- Las variables más influyentes en `cantidad_mb` son principalmente **`vlmen_fndo_ttal` y `cantidad_mb_lim`** (positivas).  
- Factores temporales (`fe_inventario`, `fe_proceso`) pueden tener correlaciones negativas leves.  
- El resto de variables numéricas muestran un efecto bajo o neutro.  

👉 Esto confirma que **volumen, límites y fechas** son relevantes para explicar la `cantidad_mb`.

#####  Selección final de variables

```python
# Usar cantidad_mb_lim como target
target_col = "cantidad_mb_lim"

data_f = data_encoded.copy()

# Filtrar columnas numéricas
numeric_cols = data_f.select_dtypes(include=['number']).columns.tolist()

# Quitar columnas constantes o casi constantes (std=0)
numeric_cols = [col for col in numeric_cols if data_f[col].std() > 0]

# Variables con baja correlación con el target
to_drop = [
    col for col in numeric_cols 
    if col != target_col and abs(data_f[col].corr(data_f[target_col])) < 0.05
]

# Crear dataset final
df_final = data_f[numeric_cols].drop(columns=to_drop)

print("Variables eliminadas:", to_drop)
print("Dimensiones finales:", df_final.shape)

```
