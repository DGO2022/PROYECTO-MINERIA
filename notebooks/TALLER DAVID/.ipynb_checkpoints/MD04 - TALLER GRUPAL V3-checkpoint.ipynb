## Fase 3. Data Preparation

### üßπ 4. Data cleaning

#### 4.1 Selecci√≥n inicial de variables claves

Queremos construir un modelo que prediga si un veh√≠culo ser√° APPROVED (1) o DISAPPROVED (0) sin usar mediciones de gases contaminantes.
Por lo tanto, debemos excluir columnas de emisiones directas y quedarnos solo con las caracter√≠sticas intr√≠nsecas del veh√≠culo.


a) Candidatas a ELIMINAR (administrativas, metadatos, fechas o con riesgo de leakage)

fecha_emision ‚Üí administrativa, no aporta a predecir.

fecha_corte ‚Üí administrativa.

id ‚Üí identificador √∫nico, sin valor predictivo.

fe_inventario ‚Üí fecha, podr√≠a correlacionarse con la salida, pero es administrativa.

fe_proceso ‚Üí tambi√©n fecha administrativa.

cantidad_mb_lim ‚Üí derivada directamente de cantidad_mb (redundante).

b) Candidatas a CONSERVAR (predictores √∫tiles)

producto ‚Üí tipo de combustible/producto (ej. "G Premium").

cantidad_mb ‚Üí cantidad medida (num√©rica, continua).

vlmen_fndo_ttal ‚Üí volumen total, variable t√©cnica importante.

ANON_CO_OSINERG ‚Üí c√≥digo anonimizado (puede representar centro/proveedor).

ANON_PTA_ABASTECIMIENTO ‚Üí punto de abastecimiento, podr√≠a influir en la distribuci√≥n.

c) Variable objetivo (target)

En tu dataset no tienes algo como "State (APPROVED/DISAPPROVED)".
Pero s√≠ podr√≠as definir una etiqueta objetivo dependiendo de lo que quieras predecir:

Ejemplo 1: Predecir si la cantidad fue ‚Äú0‚Äù o ‚Äú>0‚Äù ‚Üí variable binaria (similar a aprobado/desaprobado).

Ejemplo 2: Predecir rangos de vlmen_fndo_ttal (bajo, medio, alto).

Ejemplo 3: Si tienes en otra tabla un campo de ‚Äúestado‚Äù o ‚Äúvalidaci√≥n‚Äù, usarlo como State.

```python
# ----------------------------------------------
# üìå Preparaci√≥n del dataset para miner√≠a de datos
# ----------------------------------------------

import pandas as pd

# Cargar el archivo Excel
file_path = "Diecismuenvemil.xlsx"
data = pd.read_excel(file_path, sheet_name="Sheet1")

# ==============================
# a) Candidatas a ELIMINAR
# ==============================
eliminar = [
    'fecha_emision',    # administrativa
    'fecha_corte',      # administrativa
    'id',               # identificador √∫nico
    'fe_inventario',    # fecha administrativa
    'fe_proceso',       # fecha administrativa
    'cantidad_mb_lim'   # redundante (derivada de cantidad_mb)
]

# ==============================
# b) Candidatas a CONSERVAR
# ==============================
conservar = [
    'producto',                # tipo de combustible
    'cantidad_mb',             # cantidad medida
    'vlmen_fndo_ttal',         # volumen total
    'ANON_CO_OSINERG',         # c√≥digo anonimizado (centro/proveedor)
    'ANON_PTA_ABASTECIMIENTO'  # punto de abastecimiento
]

# ==============================
# c) Variable objetivo (target)
# ==============================
# Definimos como ejemplo: si cantidad_mb > 0 ‚Üí 1 (APROBADO), si =0 ‚Üí 0 (DESAPROBADO)
data['target'] = (data['cantidad_mb'] > 0).astype(int)

# ==============================
# Crear dataset final para modelado
# ==============================
df_modelo = data[conservar + ['target']]

print("Variables eliminadas:", eliminar)
print("Variables conservadas:", conservar)
print("\nVista previa del dataset preparado:")
df_modelo.head()
```

```python
# ==============================
# Selecci√≥n de variables relevantes + objetivo
# ==============================

# Definimos las columnas a conservar
features_keep = [
    "producto",
    "cantidad_mb",
    "vlmen_fndo_ttal",
    "ANON_CO_OSINERG",
    "ANON_PTA_ABASTECIMIENTO"
]

# Variable objetivo
target = ["target"]

# Crear nuevo dataframe solo con las columnas seleccionadas
df_sel = data[features_keep + target]

print("df_sel.shape:", df_sel.shape)
print("df_sel.columns:", df_sel.columns.tolist())

# Mostrar primeras filas
df_sel.head(3)
```

#### üìä Interpretaci√≥n de correlaciones

##### 1. Variables con correlaci√≥n positiva (asociadas a mayor `cantidad_mb`)
- **Ejemplo: `vlmen_fndo_ttal` (r ‚âà +0.xx):** cuando el volumen de fondo es mayor, tambi√©n aumenta la `cantidad_mb`.  
- **Ejemplo: `cantidad_mb_lim` (r ‚âà +0.xx):** correlaci√≥n positiva indicando que un l√≠mite m√°s alto est√° asociado con mayor cantidad.

##### 2. Variables con correlaci√≥n negativa (asociadas a menor `cantidad_mb`)
- **Ejemplo: `fe_proceso` (r ‚âà -0.xx):** fechas de proceso m√°s recientes/altas se asocian a menor cantidad.  
- **Ejemplo: `fe_inventario` (r ‚âà -0.xx):** posible efecto temporal negativo.

##### 3. Variables neutras o con efecto muy bajo
- Variables con correlaci√≥n cercana a **0** muestran poca relaci√≥n directa con la `cantidad_mb`.

---

##### üìå Conclusiones
- Las variables m√°s influyentes en `cantidad_mb` son principalmente **`vlmen_fndo_ttal` y `cantidad_mb_lim`** (positivas).  
- Factores temporales (`fe_inventario`, `fe_proceso`) pueden tener correlaciones negativas leves.  
- El resto de variables num√©ricas muestran un efecto bajo o neutro.  

üëâ Esto confirma que **volumen, l√≠mites y fechas** son relevantes para explicar la `cantidad_mb`.

#####  Selecci√≥n final de variables

```python
# Usar cantidad_mb_lim como target
target_col = "cantidad_mb_lim"

data_f = data_encoded.copy()

# Filtrar columnas num√©ricas
numeric_cols = data_f.select_dtypes(include=['number']).columns.tolist()

# Quitar columnas constantes o casi constantes (std=0)
numeric_cols = [col for col in numeric_cols if data_f[col].std() > 0]

# Variables con baja correlaci√≥n con el target
to_drop = [
    col for col in numeric_cols 
    if col != target_col and abs(data_f[col].corr(data_f[target_col])) < 0.05
]

# Crear dataset final
df_final = data_f[numeric_cols].drop(columns=to_drop)

print("Variables eliminadas:", to_drop)
print("Dimensiones finales:", df_final.shape)

```
